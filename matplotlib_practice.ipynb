```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - method 1
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - method 2
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - method 3
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Use plt.show() function
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Annotating charts: MATLAB methods
# Create the plot and add a label for the legend. Then, change line color, width, add markers.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# Create labels for the x- and y- axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y-limit between 0 and 45.
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Skill Drill
# Annotating charts: Object-Oriented Interface method
# Create plot, change line color, marker, line width, add line label for legend
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")

# Add label for chart
ax.set_title("PyBer Fare by Month")

# Add labels for axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Add y-axis limit
ax.set_ylim(0, 45)

# Add grid lines
ax.grid()

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
# Create bar chart: MATLAB methods
plt.bar(x_axis, y_axis, color="green", label="Boston")

# Create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()
```

```python
# Create horizontal bar chart: MATLAB methods
plt.barh(x_axis, y_axis)
```

```python
# Horizontal bar chart with swapped axes: MATLAB methods
plt.barh(y_axis, x_axis)
```

```python
# Original horizontal chart, y-axis order reversed: MATLAB methods
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Skill Drill: change bar color to magenta, add legend label, add title and axis labels, use inverted y-axis version: MATLAB methods
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
```

```python
# Create bar chart: obj.-oriented Method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a horizontal bar chart: obj.-oriented method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create a horizontal bar chart, switched axes: obj.-oriented method
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Skill Drill
# Create a horizontal bar chart, original axes with inverted y: obj.-oriented method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="c", label="Chicago")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.legend()
```

```python
# Create a scatterplot: MATLAB method 1
plt.plot(x_axis, y_axis, "o")
```

```python
# Create a scatterplot: MATLAB method 2
plt.scatter(x_axis, y_axis)
```

```python
# Skill Drill: make annotation changes
plt.scatter(y_axis, x_axis, c="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.xlim(0, 45)
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Create a bubble chart: MATLAB method
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Adjust bubble size by multiplying y_axis values by another value: method 1
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Adjust bubble size by multiplying y_axis values by another value: method 2 (using "list comprehension" technique)
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# Create scatter plot: obj.-oriented method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create bubble chart: obj-oriented method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Skill Drill: annotation changes
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c="skyblue", edgecolor="k", alpha=0.8, linewidth=2, label="Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0, 50)
ax.invert_yaxis()
ax.legend()
```

```python
# Create a pie chart: MATLAB method
plt.pie(y_axis, labels=x_axis)
plt.show
```

```python
# Add annotation fanciness
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%")
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
